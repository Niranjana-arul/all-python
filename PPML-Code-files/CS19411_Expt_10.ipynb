```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/Continuous data/Health parameters.csv")
df
```

```python
plt.figure(figsize=(12, 8))

x_values = [str(col) for col in df.columns]
cmap = plt.get_cmap('viridis')
colors = cmap(np.linspace(0, 1, len(df)))

# Assuming the first column contains parameter names
parameter_names = df.iloc[:, 0]  # Extract parameter names from the first column

for i in range(len(df)):
    y_values = pd.to_numeric(df.iloc[i, 1:], errors='coerce')  # Skip the first column
    plt.plot(x_values[1:], y_values, label=parameter_names.iloc[i], color=colors[i])

plt.title('Line Plot - Health Parameters')
plt.xlabel('Time')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```
