```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/healthcare_dataset.csv")
df.head()
```

```python
#Aggregation operations
print("The total billing amount is: $", df['Billing Amount'].sum())
print("")
```

```python
df.describe()
```

```python
#To display the complete number instead of scientific notation
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
print(df.agg({'Billing Amount':['sum','min','max','var','mean']}))
```

```python
Medical_condition = df.groupby(by=["Medical Condition"])
Medical_condition.first() #to view the result of formed groups
```

```python
Insurance = df.groupby(by=["Insurance Provider"])
#To view the various insurance provider groups
print("Insurance providers:\n",Insurance.first()) # shows the first entry under each group
```

```python
#To view the patients of a particular insurance
print("\nPatients of Cigna:\n", Insurance.get_group("Cigna"))
```

```python
df.groupby(["Medical Condition", "Medication"]).agg('min')
```
