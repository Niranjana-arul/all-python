```python
import pandas as pd
```

```python
demo_data = pd.DataFrame({"P_ID" : (1,2,3,4,5,6,7,8),
                          "gender" : ("male", "male", "female","female", "female", "male", "female", "male"),
                          "height" : (71,73,64,64,66,69,62,72),
                          "weight" : (175,225,130,125,165,160,115,250),
                          "visits" : (1,2,4,12,2,2,1,1),
                          "checkup" : (1,1,1,1,1,1,0,0),
                          "follow_up" : (0,0,1,2,0,0,0,0),
                          "illness" : (0,0,2,7,1,1,0,0),
                          "surgery" : (0,0,0,2,0,0,0,0),
                          "ER" : ( 0,1,0,0,0,0,1,1)})

demo_data
```

```python
lab_data = pd.DataFrame({"P_ID" : (1, 2, 4, 5, 7, 8, 9, 10),
                      "Blood glucose (mg/dL)" : (120,90,180,240,70,50,95,100),
                       "STDs" : ('N','N','N','N','N','N','Y','N'),
                       "Creatinine (mg/dL)" : (0.8,1.0, 1.2, 0.9, 1.15, 1.5, 0.7, 1.2),
                       "Ferritin (ng/mL)" : (300, 245, 249, 325, 147, 243, 278, 256),
                       "Ammonia (ug/dL)" : (42, 53, 68, 75, 69, 25, 57, 41)} )

lab_data
```

```python
#Merge the DataFrames on a common key (e.g., 'patient_id')
merged_data = pd.merge(demo_data,lab_data, on='P_ID')
merged_data
```
