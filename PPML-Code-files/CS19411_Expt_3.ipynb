Series in Pandas

```python
import pandas as pd
print("Creating a series from an array")
# Sample medical data: Patient IDs and their systolic blood pressure readings (mmHg)
patient_ids = ['P001', 'P002', 'P003', 'P004', 'P005', 'P006', 'P007', 'P008']
blood_pressure = [120, 135, 142, 128, 150, 130, 138, 145]

# Create a Pandas Series
bp_series = pd.Series(blood_pressure, index=patient_ids, name='Systolic Blood Pressure')
print("Original Blood Pressure Data:")
print(bp_series)
# Indexing
print("\nBlood Pressure of Patient 4:")
print(bp_series['P004'])

# Slicing: Access the blood pressure of multiple patients
print("\nBlood Pressure of Patients 2 to 5:")
print(bp_series['P002':'P005'])
high_bp_patients = bp_series[bp_series > 140]
print("\nPatients with High Blood Pressure (>140 mmHg):")
print(high_bp_patients)

# Replacing data using slicing
print("\nLowering Blood Pressure Readings for Patients with BP > 145 mmHg...")
bp_series[bp_series > 145] = 145
print(bp_series)
print("\nReplacing Blood Pressure for Patients P003 to P006 with 133 mmHg...")
bp_series['P003':'P006'] = 133
print(bp_series)
```

```python
#Creating a Series from a dictionary
cholesterol_data = {
    'P001': 190,
    'P002': 210,
    'P003': 180,
    'P004': 195,
    'P005': 220
}
cholesterol_series = pd.Series(cholesterol_data, name='Cholesterol Levels')
print("Cholesterol Levels:")
print(cholesterol_series)

# Creating a Series from a scalar value using an index
patients = ['P006', 'P007', 'P008', 'P009']
default_cholesterol = pd.Series(200, index=patients, name='Default Cholesterol Levels')
print("\nDefault Cholesterol Levels:")
print(default_cholesterol)

# Accessing Series attributes
print("\nCholesterol Series Attributes:")
print("Size:", cholesterol_series.size)
print("Dimensions:", cholesterol_series.ndim)
print("Shape:", cholesterol_series.shape)
print("Index:", cholesterol_series.index)

print("\nDefault Cholesterol Series Attributes:")
print("Size:", default_cholesterol.size)
print("Dimensions:", default_cholesterol.ndim)
print("Shape:", default_cholesterol.shape)
print("Index:", default_cholesterol.index)
```

```python
#data filtering - Accessing specific rows
print(data.loc[10:15]) #label-based indexing
print(data.iloc[10:15]) #position-based indexing
```

```python
# Check for missing values
missing_values = data.isna().sum()
print(missing_values)
```

```python
#data filtering - based on conditions
print("Table of Patients with Glucose levels 0:\n")
data[data['Glucose'] == 0]
```

```python
#filling 0 values
data.loc[data['Glucose'] == 0,'Glucose'] = 90.000000
data.loc[data['BloodPressure'] == 0,'BloodPressure'] = data["BloodPressure"].mean()
data.iloc[75]
```

```python
#Adding a new column
data["Age in 7 years"] = data["Age"] + 7
data.head()
```

```python
avg_age = data["Age"].mean()
print("Average age of the patients:", round(avg_age))
```

```python
# Get summary statistics
summary = data.describe()
print(summary)
```

```python
# Pregnancies vs Glucose
import matplotlib.pyplot as plt
data.plot(kind='scatter', x='Pregnancies', y='Glucose', s=32, alpha=.8, color="#c967fa")
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()
# Scatter plot of 'Age' vs 'BloodPressure'
plt.bar(data['Age'], data['Glucose'], color= "#ab801a")
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.title('Age vs Glucose')
plt.show()
```
