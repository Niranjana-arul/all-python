```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/Diabetes Missing Data.csv')
data.head()
```

```python
# Check for missing values
missing_values = data.isna().sum()
print(missing_values)
```

```python
#Remove rows with more than 2 missing values

data_cleaned = data.dropna(thresh=data.shape[1] - 2)
#data.shape[1] gives the total number of columns.

missing_values = data_cleaned.isna().sum()
print(missing_values)
```

```python
data_cleaned['Skin_Fold'] = round(data_cleaned['Skin_Fold'].fillna(data_cleaned['Skin_Fold'].mean()),2)
data_cleaned
```

```python
data_cleaned['BMI'] = round(data_cleaned['BMI'].fillna(data_cleaned['BMI'].median()),2)
data_cleaned
```
